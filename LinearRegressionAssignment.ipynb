1.	What is Regression?

```python
Regression is a supervised machine learning technique used to predict a continuous numeric output based on input features.
```

2.	What is Linear Regression?

```python
Linear Regression is a supervised machine learning algorithm used to predict a continuous numerical value by finding the best-fitting straight line between input (X) and output (y).
```

3.	When to use Linear Regression? Explain the equation of a straight line.

```python
Use linear regression when the target variable is continuous, the relationship between the variables is linear, and the dataset does not have major outliers or multicollinearity. The equation is y = mX + c, where m is the slope and c is the intercept.
```

4.	What kind of plots will you use to showcase the relationship amongst the columns?

```python
We can use these plots to show relationships among columns:

1.Scatter Plot = relationship between two numerical columns
2.Heatmap (correlation matrix) = relationship among many numerical columns
3.Bar Plot = relationship between categorical and numerical columns
4.Box Plot = compare distributions across categories
```

5.	How is the best fit line chosen?

```python
By minimizing Sum of Squared Errors (SSE) using techniques like Ordinary Least Squares or Gradient Descent.
```

6.	What is gradient descent, and why is it used? Explain the maths.

```python
An optimization algorithm that minimizes loss by updating coefficients in the opposite direction of the gradient.
```

8.	What is correlation?

9.	What is multicollinearity?

```python
When two or more independent variables are highly correlated with each other.
```

10.	How to detect multicollinearity?

```python
Using: VIF (Variance Inflation Factor)
       Correlation matrix (heatmap)
```

11.	What are the remedies for multicollinearity?

```python
Remove correlated features
Use PCA
Apply regularization (Ridge/Lasso)
```

12.	What is the R-Squared Statistics?

```python
Proportion of variance in the target variable explained by the model.
```

13.	What is an adjusted R-Squared Statistics?

```python
A modified R² that penalizes unnecessary features
```

14.	Why do we use adj R-squared?

```python
It prevents overestimation of model performance when adding irrelevant features.
```

15.	Why adj R-squared decreases when we use incompetent variables?



```python
Because adding weak predictors increases penalty and reduces model quality.
```

16.	How to interpret a Linear Regression model?

19.	What are Bias and Variance? What is Bias Variance Trade-off?

```python
Bias: Error from oversimplification
Variance: Error from sensitivity to noise
Trade-off: Find a balance that minimizes total error.
```

20.	What is the null and alternate hypothesis?

```python
Null (H₀): No relationship between variables
Alternate (H₁): Significant relationship exists
```

21.	What is multiple linear regression?

```python
A regression model with more than one independent variable.
```

22.	What is the OLS method? Derive the formulae used in the OLS method.

```python
A method that finds coefficients by minimizing the sum of squared errors.
Formula:
        β=(X^TX)^(-1)X^Ty
```

23.	What is the p-value? How does it help in feature selection?

```python
Probability that coefficients occurred by chance.
Low p‑value (<0.05) → feature is significant.
```

24.	How to handle categorical values in the data?

```python
One Hot Encoding
Label Encoding
Target Encoding
```

25.	What is regularization, and why do we need it?

```python
Regularization is a technique used to reduce overfitting by adding a penalty on large coefficients during model training. It keeps the model simpler and generalizes better to unseen data.
```

26.	Explain Ridge Regression.

```python
Ridge regression applies an L2 penalty that shrinks large coefficients but never reduces them to zero.
```

27.	Explain Lasso Regression.

```python
Lasso regression uses an L1 penalty, which can shrink some coefficients exactly to zero, making it useful for feature selection.
```

28.	Explain Elastic Net.

```python
Elastic Net combines L1 (Lasso) and L2 (Ridge) penalties, balancing coefficient shrinkage and feature selection. It is helpful when features are highly correlated.
```

29.	Why do we do a train test split?

```python
To evaluate how well the model performs on completely unseen data, preventing overfitting and ensuring the model generalizes.
```

30.	What is polynomial regression? When to use it?

```python
Polynomial regression models non-linear relationships by adding polynomial terms like x2,x3x^2, x^2,x3. It is used when scatter plots show curved patterns rather than straight lines.
```
